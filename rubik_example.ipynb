```python
from envs.rubik.utils.rubik_solver_utils import generate_problems_rubik
import envs.rubik.gym_rubik.envs.rubik_env as rubik_env
import numpy as np


env = rubik_env.RubikEnv()

state = env.reset()
env.render(state)

actions = env.get_all_actions()
for i in range(10):
    action = np.random.choice(actions)
    state, _, _, _ = env.step(state=state, action=action)
    env.render(state)


```
